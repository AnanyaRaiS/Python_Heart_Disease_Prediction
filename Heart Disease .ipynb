#### Heart Disease Distribution by Age

```python
bins = range(20, 90, 10)

dt=sns.histplot(x='Age',hue="HeartDisease",bins=bins,data=df,multiple='stack')
plt.title("Heart Disease Distribution by Age")

for bars in dt.containers:
    dt.bar_label(bars)

plt.show()
```


This histogram shows how heart disease cases (1) and non–heart disease cases (0) are distributed across age groups.
1. Very few people aged 20–30
Almost no cases of heart disease.

2. Age 30–40: Risk begins to appear
* People without heart disease are more than those with heart disease.
* Heart disease cases start to increase, but moderately.

3. Age 40–50: Sharp increase
* Heart disease count becomes significant.
* This is the starting age where heart disease becomes common.

4. Age 50–60: Highest counts
Maximum number of heart disease cases occur here.
* Both heart disease and healthy groups are around 200+
* Age 50–60 is the most critical range.

5. Age 60–70: High risk
* Total people are slightly fewer than 50–60
* Proportionally, more people have heart disease than those without in this age range.

6. Age 70–80: Decline in counts
* Fewer total individuals
* Heart disease still present in good proportion
* This shows older people remain high-risk even though sample size is smaller.

##### Analysis:
* Heart disease is strongly age-dependent.(The older a person gets, the higher the probability of heart disease.)
* Risk rises sharply after age 30
 The frequency of heart disease cases jumps from 40–50 onwards.
* Peak rik is at 50–60
 The highest number of heart disease cases and high total patients.
* High proportion of heart disease after age 60
 Even with fewer total people, more of them are heart patients.


#### Heart Disease by Chest Pain Type

```python
ax=sns.countplot(x='ChestPainType', hue='HeartDisease', data=df,palette="coolwarm")

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("ChestPainType vs Heart Disease")
plt.xlabel("ChestPainType")
plt.ylabel("Count")

plt.show()
```

This is a count plot showing how different chest pain types (ATA, NAP, ASY, TA)relate to the presence of heart disease.

1. ATA(Atypical Angina) and NAP(Non-Anginal Pain) categories mostly have no heart disease
ATA: 140 healthy vs 2 heart disease
NAP: 131 healthy vs 72 heart disease
Even though some heart disease cases exist, most people in ATA/NAP are healthy.

2. ASY (Asymptomatic) has the highest heart disease cases
* People with ASY chest pain have extremely high heart disease counts (392).
* This is the strongest indicator of heart disease.

3. TA (Typical Angina) has fewer people overall
But distribution is almost equal between healthy and heart disease.

##### Analysis:
* ASY chest pain type is the strongest predictor of heart disease.(People who do NOT show chest pain symptoms (ASY) are the most likely to have heart disease.)
* ATA chest pain is mostly found in healthy people.
This pattern shows ATA is generally not a dangerous sign.
* NAP has a moderate risk
Higher counts in healthy people but also a considerable number with heart disease.
* TA is low-frequency but balanced
This doesn’t show a strong differentiation.

This is a count plot comparing Sex (M/F) with the presence of Heart Disease.
* Heart disease is much more common in males than females.
* Males show a much higher risk of heart disease.

Reasons why men show more heart disease:
* Blood pressure
* Cholesterol
* Smoking rates and Alcohol consumption
* Stress and lifestyle‐related risks
